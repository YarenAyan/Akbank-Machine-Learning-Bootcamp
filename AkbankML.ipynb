**Model #1 Karar Ağacı**

Hem kategorik hem de sayısal verilerin bulunduğu bir veri setim olduğundan karar ağacı modelini denedim.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
steps = [('scaler', StandardScaler()),
('dec_tree', DecisionTreeClassifier())]
pipeline = Pipeline(steps)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=21)
params = {"dec_tree__criterion":['gini', 'entropy'],
         "dec_tree__max_depth":np.arange(3, 15)
         }
```

Ölçeklendirme ve karar ağacı sınıflandırıcısı içeren bir pipeline oluşturdum ve veri setini train ve test kümelerine ayırdım.

```python
from sklearn.model_selection import GridSearchCV
```

```python
for cv in range(3,10):
    cv_grid = GridSearchCV(pipeline, param_grid=params,cv=cv)
    cv_grid.fit(x_train, y_train)
    print("%d fold score: %3.3f" %(cv,cv_grid.score(x_test, y_test)))
    print("Best parameters: ", cv_grid.best_params_)
```

En iyi kriter gini, en uygun derinlik ise 4 çıktı ve model %86 başarıya ulaştı.

```python
best_tree = DecisionTreeClassifier(criterion='gini', max_depth=4)
best_tree.fit(x_train, y_train)
y_pred = best_tree.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Model düşük kaliteli şarapları tanıyamamıştır.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=21)
params = {"dec_tree__criterion":['gini', 'entropy'],
         "dec_tree__max_depth":np.arange(3, 15)
         }
for cv in range(3,6):
    cv_grid = GridSearchCV(pipeline, param_grid=params,cv=cv)
    cv_grid.fit(x_train, y_train)
    print("%d fold score: %3.2f" %(cv,cv_grid.score(x_test, y_test)))
    print("Best parameters: ", cv_grid.best_params_)
```

Üstteki sorunu çözmek için test boyutunu küçülttüm.

```python
best_tree = DecisionTreeClassifier(criterion='gini', max_depth=6)
best_tree.fit(x_train, y_train)
y_pred = best_tree.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Modelin düşük kalitede olanları sınıflandırmasındaki başarısı arttı ancak genel başarısı %82'ye düştü.

```python
del x['density']
del x['free sulfur dioxide']
del x['citric acid']
```

En az önem taşıyan özellikleri sildim.

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(data_set.iloc[: , -1])
y = pd.DataFrame(y.reshape(len(y),1))
```

```python
from imblearn.over_sampling import SMOTE
```

```python
strategy = {0:1700, 1:1700, 2:1700, 3:1700, 4:1700, 5:1700}
oversample = SMOTE(sampling_strategy=strategy)
x, y = oversample.fit_resample(x, y)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
y_train_cat = tf.keras.utils.to_categorical(y_train, 6)
y_test_cat = tf.keras.utils.to_categorical(y_test, 6)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

Modelin daha optimize çalışması için veriyi ölçeklendirdim.

```python
ann = tf.keras.models.Sequential(layers = None , name = None)
ann.add(tf.keras.layers.Input(shape = 8,))
ann.add(tf.keras.layers.Dense(units = 16 , activation = "relu" ))
ann.add(tf.keras.layers.Dense(units = 8 , activation = "relu" ))
ann.add(tf.keras.layers.Dense(units = 6 , activation = "sigmoid"))
ann.summary()
```

```python
ann.compile(optimizer = 'adam' , loss = 'categorical_crossentropy' ,metrics= ['accuracy'])
history = ann.fit(x_train, y_train_cat,  batch_size= 32, epochs = 150 , validation_data = (x_test,y_test_cat))
```

```python
plt.plot(history.history['accuracy'], label='Accuracy training data')
plt.plot(history.history['val_accuracy'], label='Accuracy validation data')
plt.legend()
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score
```

```python
y_pred = ann.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test_cat, axis=1)
accuracy = accuracy_score(y_true, y_pred_classes)
precision = precision_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')
print("Accuracy:", accuracy)
print("Precision:", precision)
print("F1 Score:", f1)
```
