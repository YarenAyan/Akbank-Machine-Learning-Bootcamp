Quality'deki veriler çoğunlukla ortalama olduğundan dağılım dengesizdir.

```python
corr_matrix = data_set.corr()
plt.figure(figsize=(20,15))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

Korelasyon matrisi oluşturdum.

```python
cor_target = abs(corr_matrix["quality"])
relevant_features = cor_target[cor_target>0.1]
relevant_features
```

Quality ile arasındaki kolerasyon değeri -0.1 ile +0.1 olan özellikleri bir değişkene atadım. Bu özellikleri tahminleyici olarak kullanacağım.

```python
to_drop = cor_target[cor_target<0.1]
to_drop
```

```python
type(to_drop)
```

```python
to_drop_frame = to_drop.to_frame()
to_drop_frame
```

```python
row_names = to_drop_frame.index
row_names_list = list(row_names)
row_names_list.append('quality')
y = data_set['quality'].values
x = data_set.drop(row_names_list, axis=1).values
```

Quality ile arasındaki korelasyon değeri 0'a yakın olan özellikleri bulup aralarında bağlantı olmadığından, veri analizine gereksiz karmaşıklık eklememek ve kullanacağım modeli daha doğru hale getirebilmek için eledim. Tahminleyici ve hedef özellikleri x ve y değişkenlerine atadım.

```python
y = [0 if value == 3 or value == 4 else 1 if value == 5 or value == 6 else 2 for value in y]
```

Elimizdeki verilerde orta kalite daha çok oldğundan 3 ve 4 puanlı şarapları düşük, 5 ve 6'yı orta, 7 ve 8'i yüksek kalite olarak değiştirip şarabın kalitesini üç farklı kategoriye ayırdım.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
steps = [('scaler', StandardScaler()),
('dec_tree', DecisionTreeClassifier())]
pipeline = Pipeline(steps)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=21)
params = {"dec_tree__criterion":['gini', 'entropy'],
         "dec_tree__max_depth":np.arange(3, 15)
         }
```

```python
from sklearn.model_selection import GridSearchCV
for cv in range(3,10):
    cv_grid = GridSearchCV(pipeline, param_grid=params,cv=cv)
    cv_grid.fit(x_train, y_train)
    print("%d fold score: %3.3f" %(cv,cv_grid.score(x_test, y_test)))
    print("Best parameters: ", cv_grid.best_params_)
```

```python
best_tree = DecisionTreeClassifier(criterion='gini', max_depth=4)
best_tree.fit(x_train, y_train)
y_pred = best_tree.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=21)
params = {"dec_tree__criterion":['gini', 'entropy'],
         "dec_tree__max_depth":np.arange(3, 15)
         }
for cv in range(3,6):
    cv_grid = GridSearchCV(pipeline, param_grid=params,cv=cv)
    cv_grid.fit(x_train, y_train)
    print("%d fold score: %3.2f" %(cv,cv_grid.score(x_test, y_test)))
    print("Best parameters: ", cv_grid.best_params_)
```

```python
best_tree = DecisionTreeClassifier(criterion='gini', max_depth=6)
best_tree.fit(x_train, y_train)
y_pred = best_tree.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn import tree
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(best_tree,
               feature_names = relevant_features[:-1],
               class_names=["0", "1", "2"],
               filled = True);
fig.savefig('dec_tree.png')
```
