Quality'deki veriler çoğunlukla ortalama olduğundan dağılım dengesizdir.

```python
corr_matrix = data_set.corr()
plt.figure(figsize=(20,15))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

Korelasyon matrisi oluşturdum.

```python
cor_target = abs(corr_matrix["quality"])
relevant_features = cor_target[cor_target>0.1]
relevant_features
```

Quality ile arasındaki kolerasyon değeri -0.1 ile +0.1 olan özellikleri bir değişkene atadım. Bu özellikleri tahminleyici olarak kullanacağım.

```python
to_drop = cor_target[cor_target<0.1]
to_drop
```

```python
type(to_drop)
```

```python
to_drop_frame = to_drop.to_frame()
to_drop_frame
```

```python
row_names = to_drop_frame.index
row_names_list = list(row_names)
row_names_list.append('quality')
y = data_set['quality'].values
x = data_set.drop(row_names_list, axis=1).values
```

Quality ile arasındaki korelasyon değeri 0'a yakın olan özellikleri bulup aralarında bağlantı olmadığından, veri analizine gereksiz karmaşıklık eklememek ve kullanacağım modeli daha doğru hale getirebilmek için eledim. Tahminleyici ve hedef özellikleri x ve y değişkenlerine atadım.

```python
y = [0 if value == 3 or value == 4 else 1 if value == 5 or value == 6 else 2 for value in y]
```

Elimizdeki verilerde orta kalite daha çok oldğundan 3 ve 4 puanlı şarapları düşük, 5 ve 6'yı orta, 7 ve 8'i yüksek kalite olarak değiştirdiğimizde daha dengeli bir dağılımla modelin öğrenmesi gereken sınıf sayısını düşürmüş oldum.
